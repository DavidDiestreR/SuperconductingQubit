```python
# imports

import scqubits as scq
from qfit import Fit
```

```python
# 1) definimos un sistema de scqubits (esto será tu "HilbertSpace")
transmon = scq.Transmon(
    EJ=30.0,
    EC=1.2,
    ng=0.3,
    ncut=31
)

# 2) lanzamos la app de qfit sobre ese sistema
fit_app = Fit(transmon)

# si estás en Jupyter debería aparecerte la interfaz (widgets)
fit_app
```
